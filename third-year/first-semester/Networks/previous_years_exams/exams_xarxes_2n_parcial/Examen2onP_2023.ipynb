 __1.- Explica la diferència entre protocols d’encaminament i “encaminables”. Posa un exemple. (1 punt)__

Els protocols d'encaminament són aquells encarregats de trobar la millor ruta per anar des de un dispositiu a un altre, minimitzant el cost, la mètrica. Generen, mantenen i treballen amb taules d'encaminament, les quals indican quin és el següent salt per anar de origen a destí.

Els protocols encaminables són utilitzats pels protocols d'encaminament per identificar els dispositius, enviar informació als diferents nodes implicats en la ruta. Són els que proporcionen l'adreça lògica que es farà servir per trobar la millor ruta origen-destí.

**2.- Explica els diferents tipus de xarxes de commutació que podem tenir. (1 punt)**

Tenim tres tipus de xarxes de commutació:
1. Commutació de circuits. Similar a la telefonia analògica. Estableix la ruta entre origen destí i un cop el destí accepta la transmissió de les dades el canal és exclusiu per aquesta transmissió fins que s'allibera.
2. Commutació de missatges. Decideix quin és el millor salt (següent pas per arribar des de origen a destí) i li envia TOTA la informació. Això es repeteix fins que arriba al destí. No és factible ja que no fragmenta la informació. Podriem estar parlant de trames de gigabytes. És, doncs, una aproximació teòrica
3. Commutació de paquets. Parteix de la idea de la commutació de missatges. Divideix el missatge en fragment -paquets- que transmet des de origen a destí. Trobem dues aproximacions al problema
    * Circiuts virtuals: On escollim el millor camí a priori entre origen i destí i enviem tots els paquets per aquest camí, sense l'exclusivitat que es tenia en el primer tipus de commutació. Diverses transmissions poden fer servir un mateix equip transmissor. Els paquets arriben ordenats i amb QoS.
    * Commutació per datagrames: Els paquets s'envien escollin en cada moment quin és el millor camí per arribar al destí. Poden arribar desordenats. No hi ha QoS.

__6.- Tenim els següents enllaços formats pels equips 1, 2 i 3, i els medis A i B.__

__L’enllaç associat al medi A i els equips 1 i 2 transmet trames de longitud L = 2000 Bytes de dades i 150 Bytes de ACK/NAK, a una velocitat de 500 kbits/s. La distància de A és de 10 km, la velocitat de propagació és de 1E8 m/s. L’enllaç associat al medi B i els equips 2 i 3 transmet trames de longitud L = 1200 Bytes de dades i 150 Bytes de ACK/NAK, a una velocitat de 1Mb/s. La distància de B és de 7 km i la velocitat de propagació és de 2E8 m/s. El valor del temps de transmissió de ACK NO és menyspreable. Seleccioneu el millor protocol i modifiqueu els paràmetres de K necessaris per evitar que es formi cap mena de coll d’ampolla, considerant que la probabilitat d’error en A és del 0,1% mentre que en B és del 0,25%. (2 punts)__

```python
import sympy as sp
sp.var("L")
sp.var("c")
sp.var("d")
sp.var("v")
sp.var("T_ACK")
sp.var("k")
t_tx = L/v
t_prop = d/c
T_total = t_tx + 2*t_prop + T_ACK
a = (2*t_prop + T_ACK/2)/t_tx
U = t_tx/(1+2*a)
U
```

```python
# Tram A
T_prop = t_prop.subs(d,1E4).subs(c,3E8)
T_tx = t_tx.subs(L, 8*2000).subs(v,0.5E6)
print(T_tx)
t_ack = 150*8/0.5E6
_a = a.subs([(T_ACK, t_ack),(t_prop, T_prop),(L, 8*2000),(v,0.5E6)])
print(_a)
print(1+2*_a)
```

```python
# Tram B 
T_prop = t_prop.subs(d,7E3).subs(c,2E8)
t_ack = 150*8/1E6
_b = a.subs([(T_ACK, t_ack), (t_prop, T_prop), (L, 8*1200), (v, 1E6)])
print(_b)
print(1+2*_b)
```

```python
# RQ Inactiva
P_errorA = 0.1/100
P_errorB = 0.25/100
U_a = (1- P_errorA)/(1+2*_a)
U_b = (1- P_errorB)/(1+2*_b)

print(U_a, U_b)
```

```python
# RQ Continua repetició selectiva

print("per K's iguals en A i B i iguals a 2: ")
print(1-P_errorA, 1-P_errorB)
```

```python
# RQ Inactiva en A i RQ Continua repetició selectiva en B
print(U_a, 1-P_errorB)
```

```python
# Go back N
print("per K's iguals en A i B i iguals a 2: ")
U_aGN = (1-P_errorA)/(1+2*_a*P_errorA)
U_bGN = (1-P_errorB)/(1+2*_a*P_errorB)
print(U_aGN, U_bGN)
```
