<div class="alert alert-success">
    <h1>Problema 1: Xocolata!</h1>
    <p>
        Considereu una rajola de xocolata de mida $W\times H$ on $W$ és l'amplada (width) i $H$ és l'alçada (height). 
        <img src="img/chocolate.png" width='40%'></img>
        Com veieu, aquesta rajola està formada per peces més petites de xocolata. 
        El nostre objectiu és separar totes les peces de la rajola amb el mínim cost possible. Cada tall vertical de la rajola té un cost diferent: $x_1,\dots,x_{w-1}$ i cada tall horitzontal també té un cost diferent: $y_1,\dots,y_{h-1}$.<br><br>
        Un cop fet un tall, la rajola queda partida i el cost del proper tall queda multiplicat pel nombre de peces diferents que es tallen. Per exemple, en l'imatge d'exemple, si primer tallem per $y_2$, a continuació tallem per $x_1$ i després tallem per $y_3$, el cost total serà: $y_2 + 2x_1 + 2y_3$.<br><br>
        Implementeu la funció <b>xocolata</b> que donats els talls verticals i horitzontals ens retorni l'ordre dels talls a realitzar, la quantitat de cada tall i el cost total de trencar la xocolata en peces individuals. Trobareu el format exacte a la capçalera de la funció.
    </p>      
</div>

````python
def xocolata(x, y):
    """
    Params
    ======
    :x: Llista de costs d'esquerra a dreta dels talls verticals.
    :y: Llista de costs de dalt a baix dels talls horitzontals.
    
    Returns
    =======
    :cuts: Una llista de tuples amb el format següent: 
           [(c1, n1), (c2, n2), ...] on:
           - c1 és el valor del primer tall
           - n1 és el nombre de talls c1 que hem de fer
           Exemple: [(3,1), (4,2), ...] vol dir que fem primer un tall de cost 3 i a continuació dos talls de cost 4.
           
    :cost: El cost total de tallar la xocolata en peces individuals.
    """
    # La intenció serà utilitzar els costs dels talls més grans primer, per tal d'evitar que
    # siguin multiplicats per un nombre de peces que es tallen massa gran. Un altre forma
    # d'expressar-ho és pensant que si fem el de cost més petit l'últim, no afectarà tant
    # que estigui multiplicat per un nombre més gran
    
    # Llista de tuples amb l'ordre de talls a realitzar i la quantitat de cada tall
    cuts = []
    
    # Cost total de trencar la xocolata en peces individuals
    cost = 0
    
    # Nombre de peces diferents que es tallen (valors per la component horitzontal [x] i per
    # la component vertical [y])
    mult_x = 1
    mult_y = 1
    
    # Iterem pel mínim nombre de talls que haurem de fer (x * y)
    n = 0
    t = len(x) * len(y)
    
    # Ordenem les llistes per tal d'anar agafant els màxims costs en l'execució
    x = sorted(x, reverse = True)
    y = sorted(y, reverse = True)
    
    # Pararem el bucle si n >= t o bé si totes dues llistes estan buides
    while (n < t) and (x or y):
        # Si encara hi ha talls x possibles, agafem el màxim
        if x:
            tall_x = x[0]
        
        # Si encara hi ha talls y possibles, agafem el màxim
        if y:
            tall_y = y[0]
             
        # Si el tall x és més gran (o igual) i encara hi ha talls possibles a la x
        if ((tall_x >= tall_y or not y) and x):
            # Eliminem de la llista un tall x possible
            x.remove(tall_x)
            
            # Fem un append a la llista amb el cost del tall i el número de peces diferents
            # que són tallades
            cuts.append((tall_x, mult_y))
            
            # Sumem al cost total el cost del tall i el numero de peces que talla
            cost += tall_x * mult_y
            
            # Cada break augmenta el nombre de peces en 1 (divideix una peça en 2)
            mult_x += 1
           
        # Si el tall y és més gran i encara hi ha talls possibles a la y
        elif((tall_x < tall_y or not x) and y):
            # Eliminem de la llista un tall y possible
            y.remove(tall_y)
            
            # Fem un append a la llista amb el cost del tall i el número de peces diferents
            # que són tallades
            cuts.append((tall_y, mult_x))
            
            # Sumem al cost total el cost del tall i el numero de peces que talla
            cost += tall_y * mult_x
            
            # Cada break augmenta el nombre de peces en 1 (divideix una peça en 2)
            mult_y += 1
            
        n += 1

    return cuts, cost
````

    <h1>Problema 2: Un alfabet estrany</h1>
    <p>
        Volem enviar missatges utilitzant el mínim d'espai possible utilitzant una codificació de dos caràcters '.' i '-'. Per fer-ho hem de traduïr cada lletra de l'alfabet català a aquesta representació. <br>
        Per exemple, podem assignar que la lletra A correspon a l'string '.', la lletra B a l'string '.-' i la lletra C a l'string '--'. Així, la paraula ABC ens quedaria codificada com '.-.--'.<br><br>
        Per a que la codificació sigui bona i reversible, és a dir, que siguem capaços de desxifrar-la, és important que els strings de codificació que triem no portin a errors.<br>
        Per exemple, si codifiquem A='.', B='-.', C='.-' i tenim l'string '.-..-', aquest pot representar tant la paraula ABC com la paraula CAC i, per tant, no és una bona codificació.
        Seguiu els següents passos per a implementar la solució:
    </p>  
    <h2>2.1 Trobar la codificació</h2>
    <ol>
        <li>Implementeu una funció <code>compute_frequency</code> que calculi la freqüència de cada lletra, és a dir, el nombre de cops que apareix a dins l'string.   
        <li>Ara crearem un arbre binari per emmagatzemar els nodes. Utilitza la classe <code>Node</code> seguint les indicacions:
        <ol>
            <li>Inicialitza una llista <code>nodes_list</code> on cada lletra sigui un node i tingui com a valor, el càlcul de freqüència del primer apartat.
            <li>Mentre la llista tingui dos nodes com a mínim, extreu de la llista els dos nodes amb la frequencia més petita i guarda'ls a l'arbre assignant-els-hi el mateix pare. 
            <li>Assigna el caràcter '.' al node de la dreta i el caràcter '-' al node de l'esquerra.
            <li>Guarda el node pare a <code>nodes_list</code> assignant-li com a valor de freqüència la suma dels seus dos fills.
            <li>Torna al punt B.
        </ol>
        <img src="img/graph.png" width='25%'></img><br>
        <li>Assigna un codi a cada fulla de l'arbre de la forma següent: sempre que descendeixis a un node de la dreta utilitza el caràcter '.' i al descendir al node de l'esquerra utilitza el caràcter '-'.
        <li>Retorna aquesta assignació com a diccionari.
    </ol>
    <h2>2.2 Codificar i decodificar</h2>
    <ol>
        <li>Crea una funció <code>encode</code> que utilitzi el diccionari retornat anteriorment per a codificar un text d'entrada
        <li>Crea una funció <code>decode</code> que utilitzi el diccionari retornat anteriorment per a decodificar un text d'entrada
        <li>Executa la funció <code>alphabet</code> i comprova que el resultat sigui l'esperat.
    </ol>
    
</div>

Exemple d'execució:

- Text a codificar: ABABAC
- Diccionari de freqüències: {'A': 3, 'B': 2, 'C': 1}
- Diccionari de conversió: {'C': '--', 'B': '-.', 'A': '.'}
- Codificat: .-..-..--
- Decodificat: ABABAC

````python
from collections import defaultdict

def compute_frequency(text):
    """
    Params
    ======
    :text: El text que volem codificar
    
    Returns
    =======
    :dct: Un diccionari amb el nombre de cops que apareix cada simbol en el text d'entrada. Per exemple: {'A': 3, 'B': 2, 'C': 1}
    """
    dct = defaultdict(int)
    
    # Iterem per cada caràcter del text a codificar i augmentem el nombre de cops que apareix
    for item in text:
        dct[item] += 1
    
    return dict(dct)
````

````python
class Node:
    """
    Aquesta classe emmagatzema la informació dels nodes de l'arbre binari.
    """
    def __init__(self, node, value, left=None, right=None):
        self.node = node    # String que representa el node
        self.value = value  # Valor de freqüència
        self.left = left    # Node de l'esquerra
        self.right = right  # Node de la dreta
        self.code = ''      # Codificació del node

    def set_code(self, code):
        self.code = code

    def __repr__(self):
        return f"Node({self.node},{self.value},{self.code})"
    
    # He afegit dos mètodes per tal de comparar instàncies de la classe Node a l'hora d'utilitzar heaps
    def __lt__(self, other):
        return (self.value < other.value)
    
    def __gt__(self, other):
        return (self.value > other.value)
````

````python
# Tal i com està implementat, codifica i decodifica correctament però amb una codificació diferent a la del exemple donat (provat amb 'ABABAC').
# Això és degut a que en el moment en què en el heap tenim dos nodes amb la mateixa freqüència, a l'hora de fer el push, els ordena arbitràriament
# i obtinc un arbre com si fos un mirall del de l'exemple. Tot i així, funciona correctament.

import heapq

def assign_codes(text, counts):   
    """
    Aquesta funció construeix el diccionari de conversió de lletres a símbols '.' i '-'.
    
    Params
    ======
    :text: El text que volem convertir
    :counts: El diccionari de freqüències que ens retorna la funció compute_frequency
    
    Returns
    =======
    :codes: El diccionari de conversió. Per exemple: {'C': '--', 'B': '-.', 'A': '.'}
    """
    # Heap de nodes on cada lletra és un node i té com a valor la freqüència calculada
    nodes_list = []
    
    # Afegim els nodes al heap segons la seva freqüència
    for key, value in counts.items():
        heapq.heappush(nodes_list, Node(key, value))
        
    # Mentre la llista tingui dos nodes com a mínim
    while len(nodes_list) > 1:
        # Extraiem els dos nodes més petits del heap
        left = heapq.heappop(nodes_list)
        right = heapq.heappop(nodes_list)
        
        # Assignem el caràcter '.' al node de la dreta i el caràcter '-' al node de l'esquerra
        left.set_code('-')
        right.set_code('.')
        
        # Guardem els nodes left i right assignant-los al pare, que tindrà la suma de les seves freqüències
        newNode = Node('', left.value+right.value,left,right)
        
        # Afegim el node pare al heap
        heapq.heappush(nodes_list, newNode)
        
    # Diccionari de conversió
    codes = {}
    
    # Node arrel de l'arbre binari
    node_arrel = heapq.heappop(nodes_list)
    
    # Funció auxiliar que recorre l'arbre binari per tal de formar el diccionari de conversió
    codificar(node_arrel, codes)
    return codes

def codificar(node, dictCodificacio, val = ''):
    codificacio = val + node.code
    
    # Recorregut inorder per recurssió
    if(node.left):
        codificar(node.left, dictCodificacio, codificacio)
        
    # Si el node és una fulla (no té fills), afegim la seva codificació
    if(not node.left and not node.right):
        dictCodificacio[node.node] = codificacio
        
    if(node.right):
        codificar(node.right, dictCodificacio, codificacio)
````

````python
def encode(text, diccionari):
    """
    Donat un text a codificar i un diccionari de conversió, codifica el text.
    
    Params
    ======
    :text: El text que volem codificar
    :diccionari: El diccionari de conversió que farem servir
    
    Returns
    =======
    :code: Una representació del text usant només els caràcters '.' i '-'
    
    """
    code = ''
    
    # Iterem per cada caràcter del text i afegim la seva codificació al string a retornar
    for lletra in text:
        code += diccionari[lletra]
    
    return code
````

````python
# Està implementat de tal manera de que si llegeix un simbol que no es troba al diccionari de conversió
# només retornarà el que hagi pogut decodificar fins trobar-se amb aquest/s caràcters

# Si és un simbol que si que està al diccionari de conversió ('-' i '.'), poden pasar dues coses:
#     1. Tot i no estar associat a cap lletra directament, es junta amb un altre caràcter i acaba formant
#        una lletra.
#     2. No aconsegueix fer cap combinació i continua iterant fins al final sense acabar de decodificar la resta.  

def decode(text, diccionari):
    """
    Donat un text a decodificar i un diccionari de conversió, decodifica el text.
    
    Params
    ======
    :text: El text que volem decodificar (caràcters '.' i '-')
    :diccionari: El diccionari de conversió que hem fet servir per codificar
    
    Returns
    =======
    :code: El text resultant de la decodificació.
    """
    
    # Dos variables que utilitzarem com a punters (per fer slicing del text)
    i = 0
    j = 1
    
    # Text resultant de la codificació
    code = ''
    
    # Invertim el diccionari per tal de poder accedir mitjançant la codificació
    inv_dict = {v: k for k, v in diccionari.items()}
    
    # Boolean que ens indica si hem saltat a l'hora d'iterar amb els punters
    # Amb saltar volem dir que a l'hora de fer un slicing (per exemple, text[0:1]), no hem trobat
    # cap codificació al diccionari amb el caràcter resultant, llavors avançem el punter j per fer
    # un altre slicing (text[0:2]), si en aquest cas sí que es troba al diccionari, llavors el 
    # següent slicing serà text[2:3], hem fet saltar la i de 0 a 2.
    saltat = False

    # Per cada simbol (caràcter) en el text, iterem 
    for simbol in text:
        # Si el fragment de text és una codificació del nostre diccionari, l'afegim al text resultant
        if inv_dict.get(text[i:j]):
            code += inv_dict.get(text[i:j])
            
            # Si per obtenir aquest slicing del text (text[i:j]) hem saltat
            if(saltat):
                i = j
                j += 1
                saltat = False
            
            # Si no hem saltat, avancem augmentant 1 cada punter
            else:
                i += 1
                j += 1
        
        # Si el resultat de fer slicing (text[i:j]) no es troba al diccionari, haurem d'avançar només
        # el punter j, i indicar que haurem de saltar (la i) quan si que trobem la codificació al diccionari
        else:
            j += 1
            saltat = True
            
    return code
````

````python
def alphabet(text):
    """
    Funció per comprovar que el codi s'executa correctament. No modifiqueu aquesta funció.
    
    Params
    ======
    :text: Missatge que volem fer servir per provar les funcions.
    """
    
    print("Text a codificar:", text)
    
    counts = compute_frequency(text)
    print("Diccionari de freqüències:", counts)
    
    codes = assign_codes(text, counts)
    print("Diccionari de conversió:", codes)
    
    codi = encode(text, codes)
    print("Codificat:", codi)
    print("Decodificat:", decode(codi, codes))
````
